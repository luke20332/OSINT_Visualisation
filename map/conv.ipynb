```python
import json
import csv

# Open the JSON file
with open('afghan.Json.txt') as f:
    data = json.load(f)

# Extract the necessary data from the JSON file
rows = []
for item in data:
    title = item['title']
    from_date = item['from']
    to_date = item['to']
    poly = item['poly']
    rows.append([title, from_date, to_date, poly])

# Write the extracted data to a CSV file
with open('afghan.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Title', 'From', 'To', 'borders'])
    writer.writerows(rows)
```

```python
import pandas as pd

# Load the JSON file into a pandas DataFrame
df = pd.read_json('afghan.Json.txt')

# Create a new DataFrame to hold the extracted titles
titles_df = pd.DataFrame(columns=['title'])

# Loop through each row of the original DataFrame
for index, row in df.iterrows():
    # Extract the title from the row
    title = row['title']
    # Add the title to the titles DataFrame
    titles_df = titles_df.append({'title': title}, ignore_index=True)

# Save the titles DataFrame as a CSV file
titles_df.to_csv('titles.csv', index=False)
```
